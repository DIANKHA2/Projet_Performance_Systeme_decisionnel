```python
from sklearn import datasets
ddd= datasets.load_iris()
```

```python
ddt = pd.DataFrame(ddd['data'],columns=ddd['feature_names'])
tar = pd.DataFrame(ddd['target'],columns=['cate'])
ddt = pd.concat([ddt,tar], axis=1)
```

```python
ddt0 = ddt[ddt['cate']==0]
ddt1 = ddt[ddt['cate']==1]
ddt2= ddt[ddt['cate']==2]
plt.scatter(ddt0["sepal length (cm)"],ddt0["petal width (cm)"], c='red')
plt.scatter(ddt1["sepal length (cm)"],ddt1["petal width (cm)"], c='yellow')
plt.scatter(ddt2["sepal length (cm)"],ddt2["petal width (cm)"], c='green')
```
